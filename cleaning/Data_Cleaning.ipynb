```python
import re

def extract_minutes(value):
    if pd.isna(value):
        return np.nan
    text = str(value).strip().lower()
    if not text or text in {'nan', 'n/a', 'none', 'no waiting time'}:
        return np.nan

    # Convert hours to minutes first, if present
    hour_match = re.search(r"(\d+(?:\.\d+)?)\s*(?:h|hr|hour|hours)", text)
    if hour_match:
        return float(hour_match.group(1)) * 60

    # Extract the first minute value; handles "10-15 min", "15 mins", "up to 20 minutes"
    minute_match = re.search(r"(\d+(?:\.\d+)?)\s*(?:m|min|mins|minute|minutes)", text)
    if minute_match:
        return float(minute_match.group(1))

    # If only a numeric range is present without units, take the average
    range_match = re.search(r"(\d+(?:\.\d+)?)\D+(\d+(?:\.\d+)?)", text)
    if range_match:
        return (float(range_match.group(1)) + float(range_match.group(2))) / 2

    # Fallback to any number present
    num_match = re.search(r"(\d+(?:\.\d+)?)", text)
    if num_match:
        return float(num_match.group(1))

    return np.nan

# Create a numeric waiting time in minutes
if 'waiting_time_min' not in df.columns:
    df['waiting_time_min'] = df['waiting_time'].apply(extract_minutes)
else:
    df['waiting_time_min'] = df['waiting_time'].apply(extract_minutes)

print("Extracted waiting_time_min as numeric minutes.")
print(df[['waiting_time', 'waiting_time_min']].head(10))
```
